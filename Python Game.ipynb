A spawn function for when the game starts or the player loses life

```python
def spawn_player():
    seed()
    sequence = [3,4,5,6,16,22]
    return choice(sequence)
```

The start of the game 

```python
lives=3
loc = spawn_player()

while lives>0:
        
    print("You find yourself in a "+re.split('[0-9]',locations[loc])[0]+'\n')
    
    if loc in death_locations.keys():
        time.sleep(1)
        print("You experience am eerie feeling!")
        time.sleep(3)
        print("...")
        time.sleep(1)
        print("Oh no!!!"+death_locations[loc])
        lives=lives-1
        print(f"You have {lives} lives left")
        loc = spawn_player()
    
    elif loc==12:
        time.sleep(1)
        print("You experience am eerie feeling!")
        time.sleep(3)
        print("...")
        time.sleep(1)
        print("Oh no!!! You see a large army coming closer!\n What could their intentions be?\n") 
        time.sleep(1)
        print("Their big and burly general calls you king and bows down!!! Is this a ploy to earn your trust???\n")
        time.sleep(1)
        print("YOU WIN!!!!!!!!!")
    
    else:
        objects_in_surrounding = {}
        count=0
        
        #Iterating
        for i in gameMap[loc]:
            if i>0:
                objects_in_surrounding[i] = count
                count+=1
                
        for i in range(1,8):
            if i in objects_in_surrounding.keys():
                print(f"In your {directions[i]}({i}) lies a {re.split('[0-9]',locations[objects_in_surrounding[i]])[0]}")
        
        user_input=""
        
        while True:
            try:
                user_input = int(input("\nWhich direction do you want to go?  "))
            
            except ValueError: 
                print("The entered value is not a number!")
                
            else:
            
                if user_input in objects_in_surrounding.keys():
                    print("You move "+ directions[user_input])
                    loc=objects_in_surrounding[user_input]
                    break
                
                else: 
                    print("Your path is blocked in that direction!")

print("GAME OVER")
```
